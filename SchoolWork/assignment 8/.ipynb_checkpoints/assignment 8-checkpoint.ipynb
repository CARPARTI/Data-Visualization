```python
ddf["View Difference"] = ddf["Total Views"] - ddf["Average Total Views"]
```

```python
ddf1 = ddf[ddf["Publish Month"] == "September"]
ddf2 = ddf[ddf["Publish Month"] == "October"]
ddf3 = ddf[ddf["Publish Month"] == "November"]
##ddf1 = ddf[(ddf["Video title"].isin(["COMP 4304 / 6934 Lecture 12","COMP 4304 / 6934 Lecture 13","COMP 4304 / 6934 Lecture 14","COMP 4304 / 6934 Lecture 15","COMP 4304 / 6934 Lecture 16","COMP 4304 / 6934 Lecture 17","COMP 4304 / 6934 Lecture 18","COMP 4304 / 6934 Lecture 19","COMP 4304 / 6934 Lecture 20","COMP 4304 / 6934 Lecture 21"]))]
```

```python
@interact(month = (["September"]) + (["October"]) +(["November"]))
          
def f(month):
    if(month=="September"):
        ax = px.line(x="Day",y="View Difference" ,data_frame=ddf1, color ="Video title")
    if(month=="October"):
        ax = px.line(x="Day",y="View Difference" ,data_frame=ddf2, color ="Video title")
    if(month=="November"):
        ax = px.line(x="Day",y="View Difference" ,data_frame=ddf3, color ="Video title")

    ax.update_xaxes(range = [0,70])
    ax.update_yaxes(range = [-60,60])

    ax.show()
        
```

## Question 3

```python
df =pd.read_csv("datavis_videos_q3.csv")
```

```python
df.head()
```

```python
df = df.groupby(["nameofday","dayofweek"]).mean().reset_index()
```

```python
df.drop_duplicates(subset="nameofday")
```

```python
sorter = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday','Sunday']
sorterIndex = dict(zip(sorter,range(len(sorter))))
sorterIndex
df['Day_id'] = df.nameofday
df['Day_id'] = df['Day_id'].map(sorterIndex)
df.sort_values('Day_id', inplace=True)
df
```

```python
fig = px.line_polar(df, r='Total Daily Views', theta='nameofday',line_close=True)
fig.update_traces(line_shape="spline")
fig.show()
```
