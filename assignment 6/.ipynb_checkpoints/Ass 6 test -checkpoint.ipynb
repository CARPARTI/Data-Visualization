```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

from ipywidgets import interact, interact_manual 
import ipywidgets as widgets
```

```python
df = pd.read_csv('Documents/phds_country.csv')

total8 = df['All Fields - Female'] + df['All Fields - Male']
df['Female'] = ((df['All Fields - Female'] / total8) * 100)
df['Male'] = ((df['All Fields - Male'] / total8) * 100)

df_female = df.sort_values(by='Female', ascending=True)
df_male = df.sort_values(by = 'Male', ascending=False)

total = df['All S&E - Female'] + df['All S&E - Male']
df['AllSEF'] = ((df['All S&E - Female'] / total) * 100)
df['AllSEM'] = ((df['All S&E - Male'] / total) * 100)

df_allsciFemale = df.sort_values(by='AllSEF' , ascending=True)
df_allsciMale = df.sort_values(by= 'AllSEM' , ascending=False)

total2 = df['Physical & Biological Sciences - Female'] + df['Physical & Biological Sciences - Male']
df['PBSF'] = ((df['Physical & Biological Sciences - Female'] / total2) * 100)
df['PBSM'] = ((df['Physical & Biological Sciences - Male'] / total2) * 100)

df_pbsFemale = df.sort_values(by='PBSF' , ascending=True)
df_pbsMale = df.sort_values(by= 'PBSM' , ascending=False)

total3 = df['Mathematics & Computer Sciences - Female'] + df['Mathematics & Computer Sciences - Male']
df['MCSF'] = ((df['Mathematics & Computer Sciences - Female'] / total3) * 100)
df['MCSM'] = ((df['Mathematics & Computer Sciences - Male'] / total3) * 100)

df_mscFemale = df.sort_values(by='MCSF' , ascending=True)
df_mcsMale = df.sort_values(by= 'MCSM' , ascending=False)

total4 = df['Agricultural Sciences - Female'] + df['Agricultural Sciences - Male']
df['ASF'] = ((df['Agricultural Sciences - Female'] / total4) * 100)
df['ASM'] = ((df['Agricultural Sciences - Male'] / total4) * 100)

df_asFemale = df.sort_values(by='ASF' , ascending=True)
df_asMale = df.sort_values(by= 'ASM' , ascending=False)

total5 = df['Social & Behavioral Sciences - Female'] + df['Social & Behavioral Sciences - Male']
df['SBSF'] = ((df['Social & Behavioral Sciences - Female'] / total5) * 100)
df['SBSM'] = ((df['Social & Behavioral Sciences - Male'] / total5) * 100)

df_sbsFemale = df.sort_values(by='SBSF' , ascending=True)
df_sbsMale = df.sort_values(by= 'SBSM' , ascending=False)

total6 = df['Engineering - Female'] + df['Engineering - Male']
df['EF'] = ((df['Engineering - Female'] / total6) * 100)
df['EM'] = ((df['Engineering - Male'] / total6) * 100)

df_eFemale = df.sort_values(by='EF' , ascending=True)
df_eMale = df.sort_values(by= 'EM' , ascending=False)

total7 = df['Non-S&E - Female'] + df['Non-S&E - Male']
df['NSEF'] = ((df['Non-S&E - Female'] / total7) * 100)
df['NSEM'] = ((df['Non-S&E - Male'] / total7) * 100)

df_nseFemale = df.sort_values(by='NSEF' , ascending=True)
df_nseMale = df.sort_values(by= 'NSEM' , ascending=False)

df
```
