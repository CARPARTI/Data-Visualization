```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns

from ipywidgets import interact, interact_manual
import ipywidgets as widgets
```

```python
df = pd.read_csv("phds_country.csv")
```

```python
df.head()
```

```python
af = df.sort_values("Region")
```

```python
df.sort_values("Region")
```

```python
df["sum"] = df["All Fields - Female"] + df["All Fields - Male"]
df["sumsize"] = (df["sum"] / df["sum"]) * 100
df["Fper"] = (df["All Fields - Female"] / df["sum"]) * 100
df["Fper"].sort_values(ascending=False)

fper = df.sort_values("Fper", ascending=False)
df["Mper"] = (df["All Fields - Male"] / df["sum"]) * 100
mper = df.sort_values("Mper", ascending=False)
```

```python
df["AllSE"] = df["All S&E - Female"] + df["All S&E - Male"]
df["FSE"] = (df["All S&E - Female"] / df["AllSE"]) * 100
df["MSE"] = (df["All S&E - Male"] / df["AllSE"]) * 100
```

```python
df["AllP&B"] = df["Physical & Biological Sciences - Female"] + df["Physical & Biological Sciences - Male"]
df["FP&B"] = (df["Physical & Biological Sciences - Female"] / df["AllP&B"]) * 100
df["MP&B"] = (df["Physical & Biological Sciences - Male"] / df["AllP&B"]) * 100
```

```python
df["AllMath"] = df["Mathematics & Computer Sciences - Female"] + df["Mathematics & Computer Sciences - Male"]
df["FMath"] = (df["Mathematics & Computer Sciences - Female"] / df["AllMath"]) * 100
df["MMath"] = (df["Mathematics & Computer Sciences - Male"] / df["AllMath"]) * 100
```

```python
df["AllAgric"] = df["Agricultural Sciences - Female"] + df["Agricultural Sciences - Male"]
df["FAgric"] = (df["Agricultural Sciences - Female"] / df["AllAgric"]) * 100
df["MAgric"] = (df["Agricultural Sciences - Male"] / df["AllAgric"]) * 100
```

```python
df["AllSocial"] = df["Social & Behavioral Sciences - Female"] + df["Social & Behavioral Sciences - Male"]
df["FSocial"] = (df["Social & Behavioral Sciences - Female"] / df["AllSocial"]) * 100
df["MSocial"] = (df["Social & Behavioral Sciences - Male"] / df["AllSocial"]) * 100
```

```python
df["AllEng"] = df["Engineering - Female"] + df["Engineering - Male"]
df["FEng"] = (df["Engineering - Female"] / df["AllEng"]) * 100
df["MEng"] = (df["Engineering - Male"] / df["AllEng"]) * 100
```

```python
df["AllNonSE"] = df["Non-S&E - Female"] + df["Non-S&E - Male"]
df["FNonSE"] = (df["Non-S&E - Female"] / df["AllNonSE"]) * 100
df["MNonSE"] = (df["Non-S&E - Male"] / df["AllNonSE"]) * 100
```
