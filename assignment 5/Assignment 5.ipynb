```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Question 1

```python
sns.scatterplot(data = df2, x = "petal_length", y = "petal_width", hue = "species" )
plt.title("Iris Petal")
```

## Question 2

```python
df2['area']=df2['petal_length']*df2['petal_width']
 
```

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.barplot(y = df2["species"], x = df2["area"], order = ["setosa", "versicolor","virginica"], data =df2["area"], ci = 95)

plt.title("Average Petal Area for Iris petal")
ax.set_xlim((0,12))
ax.set_xlabel("petal_area")
plt.show()
```
