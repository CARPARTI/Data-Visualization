## Question 2

```python
df = pd.read_csv("olympic_athletes.csv")
df = df[(df.Year >= 2000) & (df.Year <= 2016)]

df1 = df[(df.Medal == 'Gold') & (df.Sport.isin(['Swimming', 'Wrestling', 'Fencing', 'Athletics', 'Shooting', 'Cycling', 'Rowing', 'Gymnastics']))]
df_combine = df1[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_swim = df[(df.Medal == 'Gold') & (df.Sport == 'Swimming')]
df_swim = df_swim[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_wres = df[(df.Medal == 'Gold') & (df.Sport == 'Wrestling')]
df_wres = df_wres[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_fenc = df[(df.Medal == 'Gold') & (df.Sport == 'Fencing')]
df_fenc = df_fenc[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_athl = df[(df.Medal == 'Gold') & (df.Sport == 'Athletics')]
df_athl = df_athl[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_shoot = df[(df.Medal == 'Gold') & (df.Sport == 'Shooting')]
df_shoot = df_shoot[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_cycl = df[(df.Medal == 'Gold') & (df.Sport == 'Cycling')]
df_cycl = df_cycl[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_row = df[(df.Medal == 'Gold') & (df.Sport == 'Rowing')]
df_row = df_row[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()

df_gym = df[(df.Medal == 'Gold') & (df.Sport == 'Gymnastics')]
df_gym = df_gym[['ID', 'Height', 'Weight', 'Sport', 'Sex', 'Year']].drop_duplicates()
```

```python

@interact(x = widgets.Dropdown(options = ['Combined', 'Swimming', 'Wrestling', 'Fencing', 'Athletics', 'Shooting', 'Cycling', 'Rowing', 'Gymnastics'], 
                               description = 'Sport'), kde = widgets.Checkbox(options = 'Kde'), display_data = widgets.Checkbox(options = 'display_data'))

def f(x, kde, display_data):
    
        
    if(x == 'Combined'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_combine, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_combine, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_combine, hue ='Sex')
            
            
    if(x == 'Swimming'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_swim, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_swim, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_swim, hue ='Sex')
            
    if(x == 'Wrestling'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_wres, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_wres, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_wres, hue ='Sex')
    
    if(x == 'Fencing'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_fenc, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_fenc, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_fenc, hue ='Sex')
            
    if(x == 'Athletics'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_athl, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_athl, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_athl, hue ='Sex')
            
    if(x == 'Shooting'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_shoot, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_shoot, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_shoot, hue ='Sex')
            
    if(x == 'Cycling'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_cycl, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_cycl, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_cycl, hue ='Sex')
            
    if(x == 'Rowing'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_row, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_row, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_row, hue ='Sex')
    
    if(x == 'Gymnastics'):
        fig, ax = plt.subplots(figsize =(10,6))
        ax = sns.histplot(x = 'Weight', y = 'Height', data = df_gym, binwidth=(5,5), binrange = ((20,140),(130,220)), stat = 'count', cmap = 'Blues')
        if(kde):
            ax = sns.kdeplot(x = 'Weight', y = 'Height', data = df_gym, fill = True, cmap = 'Blues')
        if(display_data):
            ax = sns.scatterplot(y = 'Height', x = 'Weight', data = df_gym, hue ='Sex')
            
          
    ax.set_xlim((20, 140))
    ax.set_ylim((130, 220))
    plt.title("Gold medal winning athlete from 2000 - 2016")
```
